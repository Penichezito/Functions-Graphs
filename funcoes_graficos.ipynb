```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Função 1: Função linear
def funcao_linear(x, a=1, b=0):
  """
  Retorna o valor da função linear f(x) = ax + b.

  Args:
    x: O valor de entrada.
    a: O coeficiente angular (default: 1).
    b: O coeficiente linear (default: 0).

  Returns:
    O valor de f(x).
  """
  return a * x + b

# Gráfico da Função Linear
x_linear = np.linspace(-5, 5, 100)
y_linear = funcao_linear(x_linear)

plt.figure(figsize=(8, 5))
plt.plot(x_linear, y_linear)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Função Linear: f(x) = x')
plt.grid(True)

plt.show()
```

```python
# Função 2: Função quadrática
def funcao_quadratica(x, a=1, b=0, c=0):
  """
  Retorna o valor da função quadrática f(x) = ax^2 + bx + c.

  Args:
    x: O valor de entrada.
    a: O coeficiente de x^2 (default: 1).
    b: O coeficiente de x (default: 0).
    c: O termo constante (default: 0).

  Returns:
    O valor de f(x).
  """
  return a * x**2 + b * x + c

# Gráfico da Função Quadrática
x_quadratica = np.linspace(-5, 5, 100)
y_quadratica = funcao_quadratica(x_quadratica)

plt.figure(figsize=(8, 5))
plt.plot(x_quadratica, y_quadratica)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Função Quadrática: f(x) = x^2')
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Função 3: Função exponencial
def funcao_exponencial(x, a=1, b=1): # Altere o valor default de 'b' para 1
    """
    Retorna o valor da função exponencial f(x) = a * e^(bx).

    Args:
      x: O valor de entrada.
      a: O coeficiente multiplicativo (default: 1).
      b: O expoente (default: 1).  # Alterado para 1

    Returns:
      O valor de f(x).
    """
    return a * np.exp(b * x)

# Gráfico da Função Exponencial
x_exponencial = np.linspace(-2, 2, 100)  # Ajuste do intervalo para melhor visualização
y_exponencial = funcao_exponencial(x_exponencial, b=2) # Modifique o parâmetro 'b' para controlar a inclinação

plt.figure(figsize=(8, 5))
plt.plot(x_exponencial, y_exponencial)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Função Exponencial: f(x) = e^(2x)") # Atualize o título
plt.grid(True)
plt.ylim(0,30) # limite de y para demonstrar a exponenciação

plt.show()
```

```python
# Função 4: Função Senoidal
def funcao_senoidal(x, a=1, b=1):
    """
    Retorna o valor da função senoidal f(x) = a * sen(bx)

    Args:
      x: O valor de entrada.
      a: Amplitude (default: 1).
      b: Frequência (default: 1).

    Returns:
      O valor de f(x).
    """
    return a*np.sin(b*x)

# Gráfico da Função Senoidal
x_seno = np.linspace(0, 5 * np.pi, 100)
y_seno = funcao_senoidal(x_seno)

plt.figure(figsize=(8, 5))
plt.plot(x_seno, y_seno)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Função Senoidal: f(x) = sen(x)")
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Função 5: Função Cosseno
def funcao_cosseno(x, a=1, b=1):
    """
    Retorna o valor da função cosseno f(x) = a * cos(bx)

    Args:
      x: O valor de entrada.
      a: Amplitude (default: 1).
      b: Frequência (default: 1).

    Returns:
      O valor de f(x).
    """
    return a * np.cos(b * x)

# Gráfico da Função Cosseno
x_cosseno = np.linspace(0, 5 * np.pi, 100)
y_cosseno = funcao_cosseno(x_cosseno)

plt.figure(figsize=(8, 5))
plt.plot(x_cosseno, y_cosseno)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Função Cosseno: f(x) = cos(x)")
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ... (código anterior permanece inalterado) ...

# Função 6: Comparação Seno e Cosseno
def compara_seno_cosseno(x, a=1, b=1):
    """
    Compara as funções seno e cosseno.

    Args:
      x: O valor de entrada.
      a: Amplitude (default: 1).
      b: Frequência (default: 1).

    Returns:
      Uma tupla contendo os valores do seno e cosseno de x.
    """
    seno = a * np.sin(b * x)
    cosseno = a * np.cos(b * x)
    return seno, cosseno

# Gráfico de Comparação
x_compara = np.linspace(0, 4 * np.pi, 200)  # Aumenta a resolução para melhor visualização
y_seno, y_cosseno = compara_seno_cosseno(x_compara)

plt.figure(figsize=(10, 6))
plt.plot(x_compara, y_seno, label='Seno')
plt.plot(x_compara, y_cosseno, label='Cosseno')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Comparação entre Seno e Cosseno")
plt.grid(True)
plt.legend()  # Adiciona a legenda para identificar as curvas

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Função 6: Função Tangente
def funcao_tangente(x):
    """
    Retorna o valor da função tangente f(x) = tan(x)

    Args:
      x: O valor de entrada.

    Returns:
      O valor de f(x).
    """
    return np.tan(x)

# Gráfico da Função Tangente
x_tangente = np.linspace(-np.pi/2 + 0.1, np.pi/2 - 0.1, 100) # Evitando as assíntotas
y_tangente = funcao_tangente(x_tangente)

plt.figure(figsize=(8, 5))
plt.plot(x_tangente, y_tangente)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Função Tangente: f(x) = tan(x)")
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Função 7: Sigmoid
def funcao_sigmoid(x):
    """
    Retorna o valor da função sigmoide f(x) = 1 / (1 + e^(-x)).

    Args:
      x: O valor de entrada.

    Returns:
      O valor de f(x).
    """
    return 1 / (1 + np.exp(-x))

# Gráfico da Função Sigmoide
x_sigmoid = np.linspace(-10, 10, 100)
y_sigmoid = funcao_sigmoid(x_sigmoid)

plt.figure(figsize=(8, 5))
plt.plot(x_sigmoid, y_sigmoid)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Função Sigmoide: f(x) = 1 / (1 + e^(-x))")
plt.grid(True)

plt.show()
```
